### Assessing Data

```python
def assess_data(folder_path):
    """
    Fungsi untuk membaca dan mengevaluasi semua dataset dalam folder tertentu.
    """
    # List semua file dalam folder
    files = [f for f in os.listdir(folder_path) if f.endswith(('.csv', '.xlsx'))]

    for file in files:
        # Cek apakah file CSV atau Excel
        file_path = os.path.join(folder_path, file)
        if file.endswith('.csv'):
            df = pd.read_csv(file_path)
        elif file.endswith('.xlsx'):
            df = pd.read_excel(file_path)

        # Tampilkan informasi dasar tentang dataset
        print(f"\n📂 Assessing: {file}")
        print("-" * 50)
        print(f"Shape (Rows, Columns): {df.shape}")
        print("\n🔹 Info Dataset:")
        print(df.info())

        # Cek missing values
        print("\n🔹 Missing Values:")
        print(df.isnull().sum())

        # Cek duplikasi
        print(f"\n🔹 Duplicate Rows: {df.duplicated().sum()}")
        print("=" * 50)

# Jalankan fungsi dengan path folder dataset
assess_data("data")
```

```python
# Konversi ke datetime sebelum perhitungan
orders['order_purchase_timestamp'] = pd.to_datetime(orders['order_purchase_timestamp'])
orders['order_approved_at'] = pd.to_datetime(orders['order_approved_at'])

# Hitung rata-rata delay approval (waktu dalam detik)
avg_approval_delay = (orders['order_approved_at'] - orders['order_purchase_timestamp']).dt.total_seconds().mean()

# Isi missing values dengan order_purchase_timestamp + rata-rata delay
orders['order_approved_at'] = orders['order_approved_at'].fillna(orders['order_purchase_timestamp'] + pd.to_timedelta(avg_approval_delay, unit='s'))
```

```python
def fill_category_by_name_length(df):
    # Mode global (kategori paling sering muncul di seluruh dataset)
    global_mode = df['product_category_name'].mode()[0]

    # Mengisi berdasarkan kelompok panjang nama produk
    df['product_category_name'] = df['product_category_name'].fillna(
        df.groupby('product_name_lenght')['product_category_name'].transform(lambda x: x.mode()[0] if not x.mode().empty else global_mode)
    )

    return df

# Terapkan fungsi pada dataset products
products = fill_category_by_name_length(products)

# Cek apakah masih ada missing values
print(products['product_category_name'].isnull().sum())
```

#### RFM Analysis

```python
import pandas as pd

# Merge order data dengan customer data untuk mendapatkan kota
orders_customers_payments = data_merger

# Format tanggal untuk recency
orders_customers_payments['order_purchase_timestamp'] = pd.to_datetime(orders_customers_payments['order_purchase_timestamp'])

# Menentukan snapshot date untuk menghitung recency
snapshot_date = orders_customers_payments['order_purchase_timestamp'].max() + pd.Timedelta(days=1)

# Menghitung RFM per kota
rfm_city = orders_customers_payments.groupby('customer_city').agg({
    'order_purchase_timestamp': lambda x: (snapshot_date - x.max()).days,  # Recency
    'order_id': 'count',  # Frequency
    'payment_value': 'sum'  # Monetary
}).reset_index()

# Mengganti nama kolom agar lebih jelas
rfm_city.columns = ['customer_city', 'Recency', 'Frequency', 'Monetary']

# Membuat kuantil untuk segmentasi kota
rfm_city['R_Score'] = pd.qcut(rfm_city['Recency'], 4, labels=[4, 3, 2, 1])
rfm_city['F_Score'] = pd.qcut(rfm_city['Frequency'].rank(method="first"), 4, labels=[1, 2, 3, 4])
rfm_city['M_Score'] = pd.qcut(rfm_city['Monetary'].rank(method="first"), 4, labels=[1, 2, 3, 4])

# Membuat skor RFM total
rfm_city['RFM_Score'] = rfm_city[['R_Score', 'F_Score', 'M_Score']].sum(axis=1)
rfm_city
```

```python
# Mengambil data RFM yang sudah dihitung sebelumnya
rfm_top_cities = rfm_city.sort_values(by="RFM_Score", ascending=False).head(10)

# Membuat figure dengan 3 subplot
fig, axes = plt.subplots(1, 3, figsize=(18, 5))

# Recency
sns.barplot(data=rfm_top_cities, x="customer_city", y="Recency", ax=axes[0], palette="Blues_r")
axes[0].set_title("By Recency (days)")
axes[0].set_xticklabels(axes[0].get_xticklabels(), rotation=90)

# Frequency
sns.barplot(data=rfm_top_cities, x="customer_city", y="Frequency", ax=axes[1], palette="Blues_r")
axes[1].set_title("By Frequency")
axes[1].set_xticklabels(axes[1].get_xticklabels(), rotation=90)

# Monetary
sns.barplot(data=rfm_top_cities, x="customer_city", y="Monetary", ax=axes[2], palette="Blues_r")
axes[2].set_title("By Monetary")
axes[2].set_xticklabels(axes[2].get_xticklabels(), rotation=90)

# Menampilkan plot
plt.suptitle("Best Cities Based on RFM Parameters", fontsize=16)
plt.tight_layout()
plt.show()
```

- ⏱️ **Recency:** Pelanggan di *Vitória da Conquista* dan *Vitória* juga baru-baru ini melakukan pembelian, menandakan loyalitas yang tinggi.

- 💰 **Monetary:** Kota *Vitória* mendominasi dalam total pengeluaran pelanggan, jauh di atas kota lainnya.

**💡 Interpretasi:**
Kota-kota tersebut dapat dikategorikan sebagai **"Kota Super Aktif"**, karena memiliki skor tinggi di ketiga aspek RFM.

**✅ Rekomendasi Bisnis:**
- Fokuskan promosi eksklusif atau program loyalti di kota *Vitória* dan *Vitória da Conquista*.
- Pertimbangkan membuka gudang atau mempercepat pengiriman di wilayah tersebut untuk meningkatkan pengalaman pelanggan.

```python
# Membagi nilai RFM menjadi 4 kategori (1-4)
rfm_city['R_Score'] = pd.qcut(rfm_city['Recency'], 4, labels=[4, 3, 2, 1])  # Recency: Semakin kecil, semakin loyal
rfm_city['F_Score'] = pd.qcut(rfm_city['Frequency'].rank(method="first"), 4, labels=[1, 2, 3, 4])  # Frequency: Semakin besar, semakin loyal
rfm_city['M_Score'] = pd.qcut(rfm_city['Monetary'].rank(method="first"), 4, labels=[1, 2, 3, 4])  # Monetary: Semakin besar, semakin loyal

# Menambahkan total RFM Score
rfm_city['RFM_Score'] = rfm_city[['R_Score', 'F_Score', 'M_Score']].sum(axis=1)

# Menentukan segmen berdasarkan skor RFM
def assign_segment(x):
    if x >= 10:
        return "Kota Super Aktif"
    elif x >= 7:
        return "Kota Potensial"
    elif x >= 5:
        return "Kota Perlu Perhatian"
    else:
        return "Kota Berisiko Churn"

rfm_city['Segment'] = rfm_city['RFM_Score'].apply(assign_segment)

rfm_segment = rfm_city[['customer_city', 'Segment']].rename(columns={'Segment': 'segment_kota'})
data_merger = data_merger.merge(rfm_segment, on='customer_city', how='left')

display(rfm_city)
display(data_merger)
```

```python
import geopandas as gpd
from shapely.geometry import Point
import matplotlib.pyplot as plt

# 🔄 Ambil rata-rata koordinat per kota dan rename kolomnya secara eksplisit
geo_avg = geolocation.groupby('geolocation_city')[['geolocation_lat', 'geolocation_lng']].mean().reset_index()
geo_avg = geo_avg.rename(columns={
    'geolocation_city': 'customer_city',
    'geolocation_lat': 'lat',
    'geolocation_lng': 'lon'
})

# 🔁 Gabungkan data koordinat ke data_merger yang sudah punya segmentasi
data_merger = data_merger.merge(geo_avg, on='customer_city', how='inner')

# 🧭 Buat kolom geometry dari koordinat lon/lat
geometry = [Point(xy) for xy in zip(data_merger['lon'], data_merger['lat'])]

# 🌍 Buat GeoDataFrame
gdf = gpd.GeoDataFrame(data_merger, geometry=geometry, crs="EPSG:4326")

# 🗺️ Load shapefile negara Brazil
world = gpd.read_file(".\\dashboard\\ne_110m_admin_0_countries\\ne_110m_admin_0_countries.shp")
brazil = world[world['ADMIN'] == 'Brazil']

# 🧼 Filter hanya titik-titik di dalam batas wilayah Brazil
gdf_brazil = gdf[gdf.within(brazil.unary_union)]

# 🎨 Plot segmentasi kota berdasarkan RFM
fig, ax = plt.subplots(figsize=(12, 10))
brazil.plot(ax=ax, color='lightgray', edgecolor='white')

gdf_brazil.plot(
    ax=ax,
    column='segment_kota',     # pastikan ini nama kolom di dataframe kamu
    legend=True,
    markersize=35,
    alpha=0.7,
    cmap='Set2'
)

plt.title("📍 Heatmap Segmentasi Kota berdasarkan RFM (GeoPandas)", fontsize=14)
plt.axis("off")
plt.tight_layout()
plt.show()
```
